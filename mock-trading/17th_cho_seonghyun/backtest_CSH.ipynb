```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt



codes = [
"267290","005740","058850","010100","002460",
"016710","016740","007210","004840","013520",
"002450","034590","023800","008260","013570",
"227840","248170","012320","014580","000230"
]

tickers = [code + ".KS" for code in codes]
etf = "226980.KS"



data = yf.download(tickers + [etf], period="5y")["Close"]



returns = data / data.iloc[0]

portfolio = returns[tickers].mean(axis=1)
etf_return = returns[etf]



plt.figure(figsize=(12,6))

plt.plot(portfolio.index, portfolio, label="My Portfolio", linewidth=2)
plt.plot(etf_return.index, etf_return, label="KODEX 200 중소형", linestyle='--')

plt.title("Portfolio vs KODEX 200 중소형 (5Y)")
plt.xlabel("Date")
plt.ylabel("Return (Start = 1)")
plt.legend()
plt.grid()

plt.show()



print(" 포트폴리오 수익률:", portfolio.iloc[-1])
print(" ETF 수익률:", etf_return.iloc[-1])
```
